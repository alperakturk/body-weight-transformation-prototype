```python
INPUT_NAME = "yasli.jpg"
```

```python
from rembg import remove
from PIL import Image
import numpy as np

inp = Image.open(INPUT_NAME).convert("RGB")
out = remove(inp)
out.save("segmentation.png")

rgba_np = np.array(out)
mask = (rgba_np[..., 3] > 0).astype(np.uint8)
print("mask ok:", mask.shape, "foreground:", int(mask.sum()))
```

```python
import cv2
import numpy as np

rgba = cv2.imread("segmentation.png", cv2.IMREAD_UNCHANGED)
h, w = rgba.shape[:2]

# Bel kalça göğüs vb. için uygun olduğunu düşündüğüm yükseklikler
y1 = int(h * 0.30)
y2 = int(h * 0.85)

def warp_band(rgba_img, scale_x=1.2):
    out_img = rgba_img.copy()
    band = out_img[y1:y2, :, :]

    bh, bw = band.shape[:2]
    new_w = max(1, int(bw * scale_x))
    band_scaled = cv2.resize(band, (new_w, bh), interpolation=cv2.INTER_LINEAR)

    if new_w > bw:
        start = (new_w - bw) // 2
        band_scaled = band_scaled[:, start:start+bw, :]
    else:
        pad = bw - new_w
        left_pad = pad // 2
        right_pad = pad - left_pad
        band_scaled = cv2.copyMakeBorder(
            band_scaled, 0, 0, left_pad, right_pad,
            cv2.BORDER_CONSTANT, value=(0,0,0,0)
        )

    out_img[y1:y2, :, :] = band_scaled
    return out_img

fat_rgba  = warp_band(rgba, scale_x=1.25)  # kilolu göstermek için
thin_rgba = warp_band(rgba, scale_x=0.75)  # zayıf göstermek için

cv2.imwrite("fat.png", fat_rgba)
cv2.imwrite("thin.png", thin_rgba)

print("Saved fat.png & thin.png")
```

```python
import cv2
import numpy as np

rgba = cv2.imread("segmentation.png", cv2.IMREAD_UNCHANGED)
h, w = rgba.shape[:2]

base = cv2.imread(INPUT_NAME, cv2.IMREAD_COLOR)
base = cv2.resize(base, (w, h))

def composite(base_bgr, rgba_img, blur=31):
    alpha = rgba_img[..., 3].astype(np.float32) / 255.0
    alpha = cv2.GaussianBlur(alpha, (blur, blur), 0)
    alpha = alpha[..., None]

    fg_rgb = rgba_img[..., :3].astype(np.float32)
    fg_bgr = fg_rgb[..., ::-1]  # RGB->BGR

    out = base_bgr.astype(np.float32) * (1 - alpha) + fg_bgr * alpha
    return out.astype(np.uint8)

fat_rgba  = cv2.imread("fat.png",  cv2.IMREAD_UNCHANGED)
thin_rgba = cv2.imread("thin.png", cv2.IMREAD_UNCHANGED)

final_fat  = composite(base, fat_rgba, blur=31)
final_thin = composite(base, thin_rgba, blur=31)

cv2.imwrite("final_fat.png", final_fat)
cv2.imwrite("final_thin.png", final_thin)

print("Saved final_fat.png & final_thin.png")
```

```python
import cv2
import numpy as np
from PIL import Image
from IPython.display import display

def to_rgb(img_bgr):
    return cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

seg_rgba = cv2.imread("segmentation.png", cv2.IMREAD_UNCHANGED)
h, w = seg_rgba.shape[:2]

orig = cv2.imread(INPUT_NAME)
orig = cv2.resize(orig, (w, h))
orig = to_rgb(orig)

seg_rgb = cv2.cvtColor(seg_rgba[..., :3], cv2.COLOR_BGR2RGB)

ffat = to_rgb(cv2.imread("final_fat.png"))
fthin = to_rgb(cv2.imread("final_thin.png"))

def fit(x):
    return cv2.resize(x, (w//2, h//2))

grid = np.concatenate([
    np.concatenate([fit(orig), fit(seg_rgb)], axis=1),
    np.concatenate([fit(ffat), fit(fthin)], axis=1)
], axis=0)

display(Image.fromarray(grid))
```
